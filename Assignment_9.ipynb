### <u>Logistic Regression</u>

```python
# Importing the required libraries.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

- Import datasets

```python
data = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')
data1 = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv')
```

- Concatenate the datasets

```python
result = pd.concat([data, data1])
```

```python
result.head()
```

```python
result.isnull().sum()
```

- Remove NaN values from the column ‘Survived’. 

```python
a = result.dropna(subset=['Survived'])
a.head()
```

```python
a.isnull().sum()
```

- Create a new column 'Has_Cabin'

```python
result['Has_Cabin'] = np.where(result['Cabin'].isnull(), 'yes', 'no') 
```

```python
result.head()
```

- Create a new column “Title”

```python
b=result['Title'] = result.Name.str.extract(r',\s*([^\.]*)\s*\.', expand=False)
```

```python
b.head()
```

```python
result.head()
```

```python
result = result.fillna(result.mean())
result = result.fillna(result.mode)
result.isnull().sum()
```

-  Create two new columns “Categorical_age” and “Categorical_fare”.

```python
result['Categorical_age']=pd.qcut(result['Age'],q=3,labels=False)
result['Categorical_fare']=pd.qcut(result['Fare'],q=3,labels=False)
result.head()
```

```python
c=result.drop(['Cabin','Name','PassengerId','Ticket', 'Age', 'Fare', 'SibSp', 'Parch'],axis=1)
c.head()
```

- Transform all variables into numeric ones.

```python
d = pd.get_dummies(result)
d.head()
```

```python
e = d.fillna(d.min())
e.head()
```

```python
X = e.drop(['Survived'],axis=1)
y = e['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

- Use logistic regression

```python
reg = LogisticRegression()
reg.fit(X_train,y_train)
```

```python
pred = reg.predict(X_test)
```

```python
F_rslt = pd.DataFrame({"Predicted":pred,"Actual":y_test})
```

```python
F_rslt.head()
```

-  Accuracy score.

```python
reg.score(X_test,y_test)*100
```
