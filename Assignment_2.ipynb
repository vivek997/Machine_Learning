```python
#importing numpy
import numpy as np
```

```python
# question 1
rd = np.random.rand(10,1)
print(rd)
print(rd.shape)
print(rd.size)
mean = np.mean(rd)
print(mean)
```

```python
# question 2
b = np.random.rand(20,1)
print(b)
print(b.shape)
print(b.size)
varis = np.var(b)
print(varis)
deviation = np.std(b)
print(deviation)
```

```python
# question 3
arrayA = np.random.rand(10,20)
arrayB = np.random.rand(20,25)
multi = np.matmul(arrayA,arrayB)
print(multi)
print(multi.shape)
print(np.sum(multi))
```

```python
# question 4
array = []
A = np.random.rand(10,1)
for ele in np.nditer(A):
  ele = 1 / (1 + np.exp(-ele))
  array.append(ele)
arr = np.asarray(array).reshape(10,1)
print(arr)
```
